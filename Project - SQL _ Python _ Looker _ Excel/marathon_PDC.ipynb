```python
import pandas as pd
from datetime import timedelta
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from geopy.extra.rate_limiter import RateLimiter
```

```python
df = pd.read_csv('marathon.csv')
```

```python
df.head()
```

```python
df2 = df.dropna(axis = 1)
```

```python
df2.head()
```

```python
df2['fullname'] = df2['First'] + ' ' + df2['Last']
```

```python
df2.head()
```

```python
df2['Time']  = pd.to_timedelta(df2['Time'])
```

```python
df2['Total_Minutes'] = df2['Time'].dt.total_seconds() / 60
```

```python
df2['Total_Minutes'] = df2['Total_Minutes'].round().astype(int)
```

```python
df2.head()
```

```python
df2.rename(columns = {'Division' : 'Gender'}, inplace = True)
```

```python
def get_lat_lomg(city, state):
    address = f"{city}, {state}"
    try:
        geolocator = Nominatim(user_agent="running", timeout = 10)
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        return None, None
```

```python
df2['latitude'], df2['longitude'] = zip(*df2.apply(lambda x : get_lat_lomg(x['City'], x['State']),axis = 1))
```

```python
df2.head()
```

```python
df2['latlong'] = df2['latitude'].astype(str) + ', ' + df2['longitude'].astype(str)
```

```python
df2.head()
```

```python
df2.to_csv('marathon_PDC.Csv', index = False)
```
