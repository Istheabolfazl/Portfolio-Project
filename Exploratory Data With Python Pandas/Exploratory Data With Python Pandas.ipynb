```python
# Seasons for th data -> slower in summer than winter?

# spring 3-5
# Summer 6-8
# fall 9-11
#  winter 12 - 2

# split between twe decimals
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_seasons'] = df3['race_month'].apply(lambda x : 'Winter' if x > 11 else "Fall" if x > 8 else 'Summer' if x > 5  else 'Spring')
```

```python
df3.head(25)
```

```python
df3.groupby('race_seasons')['athlete_average_spead'].agg(["mean","count"]).sort_values('mean', ascending=False)
```
